```python
# Extract data (run once)
import zipfile
import os

base_path = "/content/drive/MyDrive/"
output_folder = "/content/HAM10000_images/"

os.makedirs(output_folder, exist_ok=True)

for z in ["HAM10000_images_part_1.zip", "HAM10000_images_part_2.zip"]:
    zip_path = os.path.join(base_path, z)
    if os.path.exists(zip_path):
        print(f"Extracting {z}...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(output_folder)

print(f"Total images: {len(os.listdir(output_folder))}")
```

## 4. View Results

```python
# Display generated images
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_generated(epoch):
    img = mpimg.imread(f"/content/gan_outputs/epoch_{epoch:03d}.png")
    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.title(f"Epoch {epoch}")
    plt.axis('off')
    plt.show()

# Show progress
for e in [1, 25, 50, 75, 100]:
    try:
        show_generated(e)
    except:
        pass
```
